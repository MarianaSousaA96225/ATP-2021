### Resumo:

Recorrendo à linguagem de programação _Python_ foi resolvido o problema proposto para este trabalho. Em síntese, foi solicitado um programa que permita ao utilizador gerar listas, manual ou aleatoriamente, e de seguida, as ordenar.

Para tal, foi elaborada uma principal função _programa()_ que perguntará sucessivamente a opção que o utilizador pretende escolher (dentro das possibilidades listadas no _menu()_) , e, consoante o introduzido, irá chamar as respetivas funções.

A opção 1 permitirá gerar manualmente uma lista de um número de elementos à escolha do utilizador, chamando para tal a função _criarListaIn()_. De semelhante modo, a função _criarListaA()_ será chamada para a opção 2 para gerar uma lista aleatoriamente. Também se contra no programa a função _bubbleSort(L)_ que recebe a lista _L_ gerada previamente e retorna a lista ordenada. 
É possível ainda ver a lista corrente, limpar a lista e sair do programa.

```python
def menu():
    print ("""
    (1) Gerar uma lista manualmente;
    (2) Gerar uma lista de valores aleatórios; 
    (3) Escrever a lista corrente no stdout; 
    (4) Limpar a lista;
    (5) Ordenar a lista;
    (0) Sair 
    """)
    
def criarListaIn():
    lista = []
    n = int(input("introduza nº de elementos na lista: "))
    j = 1
    for i in range(n):
        elemento = int(input("introduza o elemento " + str(j) + " : "))
        lista.append(elemento)
        i = i + 1
        j = j + 1
    return lista

def criarListaA():
    from random import randrange
    lista = []
    n = int(input("introduza nº de elementos na lista: "))
    for i in range(n):
        elemento = randrange(0, 101)
        lista.append(elemento)
        i = i + 1
    return lista

def bubbleSort(L):
    trocas = False
    
    for i in range(len(L)) :
        if i < (len(L) - 1) :
            if L[i] > L[i + 1] :
                L[i], L[i + 1] = L[i + 1], L[i]
                trocas = True
                
    if trocas == True :
        (bubbleSort(L))
    
    return L
```

```python
def programa():
    menu()
    L = []
    opção = -1
    while opção != 0 :
        opção = int(input("introduza uma opção: "))
        if opção == 1 :
            L = criarListaIn()
        elif opção == 2 :
            L = criarListaA()
        elif opção == 3 :
            print (L)
        elif opção == 4 :
            L.clear()
        elif opção == 5 :
            L = bubbleSort(L)
        elif opção == 0 :
            print ("Até à próxima!")
        else :
            print ("opção inválida")
            
programa()
```
