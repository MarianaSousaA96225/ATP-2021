### Resumo:

Neste ficheiro encontra-se o código fonte, desenvolvido em linguagem _Python_ , que permite ao utilizidar efetuar uma variedade de __operações sobre funções__.

Inicialmente, será exibido o menu de opções ao utilizador, com recurso à _função menu()_ , a partir do qual poderá escolher a operação desejada, recursivamente, até escolher a __opção 0__ para __sair__ do programa.

Ao escolher a __opção 1__ poderá criar a fração, introduzindo o numerador e denominador na _função criarFracao()_. De seguida, a __opção 2__ recorre à _função simplificarFracao()_ para encontrar o __máximo divisor comum__ de modo a devolver a __fração simplificada__, ou seja, na sua forma mais reduzida, ao utilizador.
As __opções 3, 4, 5 e 6__ realizam as principais operações sobre as funções - recebem duas frações e retornam a sua __soma__ ( _função somarFracao()_ ) , __subtração__ ( _função subtrairFracao()_ ) , __multiplicação__ ( _função multiplicarFracao()_ ) e __divisão__ ( _função dividirFracao()_ ) , respetivamente.

Se o utilizador pretender manipular mais de duas frações, a __opção 7__ permite gerar uma lista de frações usando a _função criarListaFrac()_. As __opções 8 e 9__ irão receber a lista criada e retornam a sua __soma e multiplicação total__, utilizando para tal as _funções somarLista() e multiplicarLista()_ , respetivamente.

```python
def menu():
    print ("""
    (1) Criar fração;
    (2) Simplificar fração;
    (3) Somar frações;
    (4) Subtrair frações;
    (5) Multiplicar frações;
    (6) Dividir frações;
    (7) Criar lista de n frações;
    (8) Somar lista;
    (9) Multiplicar lista;
    (0) Sair
    """)
```

```python
def criarFracao():
    num = int(input("Introduza o numerador: "))
    den = int(input("Introduza o denominador: "))
    f = (num, den)
    return f

def criarListaFrac(n):
    listaFrac = []
    j = 1
    for i in range(n):
        num = int(input("Introduza o numerador da fração " + str(j) + ": "))
        den = int(input("Introduza o denominador da fraçao " + str(j) + ": "))
        frac = (num, den)
        listaFrac.append(frac)
        j = j + 1
        i = i + 1
    return listaFrac
```

```python
def mdc(a, b):
    res = 1
    if a > b :
        a, b = b, a
    for i in range(2, a+1) :
        if (a % i == 0) and (b % i == 0) :
            res = i
    return res

def mdc2(a, b):
    if a < b :
        return mdc(a, b)
    elif a % b == 0 :
        return b
    else :
        return mdc(a, a%b)
    
def simplificarFracao(f):
    num, den = f
    m = mdc2(num, den)
    return (int(num/m), int(den/m))
```

```python
def somarFracao(f1, f2):
    num1, den1 = f1
    num2, den2 = f2
    return (num1*den2 + num2*den1, den1*den2)

def subtrairFracao(f1, f2):
    num1, den1 = f1
    num2, den2 = f2
    return (num1*den2 - num2*den1, den1*den2)

def multiplicarFracao(f1, f2):
    num1, den1 = f1
    num2, den2 = f2
    return (num1*num2, den1*den2)

def dividirFracao(f1, f2):
    num1, den1 = f1
    num2, den2 = f2
    return (num1*den2, den1*num2)
```

```python
def somarLista(listaFrac):
    soma = listaFrac[0]
    for i in range(1, len(listaFrac)) :
        soma = somarFracao(soma, listaFrac[i])
    return soma

def multiplicarLista(listaFrac):
    mult = listaFrac[0]
    for i in range(1, len(listaFrac)) :
        mult = multiplicarFracao(mult, listaFrac[i])
    return mult
```

```python
def verFracao(f):
    num, den = f
    print (str(num) + "/" + str(den))
```

```python
def operacoesFracoes():
    opcao = -1
    while opcao != 0 :
        menu()
        opcao = int(input("Introduza uma opção: "))
        
        if opcao == 1 :
            F = criarFracao()
            verFracao(F)
        
        elif opcao == 2 :
            verFracao(simplificarFracao(F))
        
        elif opcao == 3 :
            L = criarListaFrac(2)
            F = somarFracao(L[0], L[1])
            verFracao(L[0]) 
            print("+", end = " ") 
            verFracao(L[1]) 
            print("=", end = " ") 
            verFracao(simplificarFracao(F))
        
        elif opcao == 4 :
            L = criarListaFrac(2)
            F = subtrairFracao(L[0], L[1])
            verFracao(L[0]) 
            print("-", end = " ") 
            verFracao(L[1]) 
            print("=", end = " ") 
            verFracao(simplificarFracao(F))
        
        elif opcao == 5 :
            L = criarListaFrac(2)
            F = multiplicarFracao(L[0], L[1])
            verFracao(L[0]) 
            print("*", end = " ") 
            verFracao(L[1]) 
            print("=", end = " ") 
            verFracao(simplificarFracao(F))
    
        elif opcao == 6 :
            L = criarListaFrac(2)
            F = dividirFracao(L[0], L[1])
            verFracao(L[0]) 
            print("/", end = " ") 
            verFracao(L[1]) 
            print("=", end = " ") 
            verFracao(simplificarFracao(F))
        
        elif opcao == 7 :
            n = int(input("Introduza o nº de frações: "))
            L = criarListaFrac(n)
            print (L)
        
        elif opcao == 8 :
            F = somarLista(L)
            verFracao(F)
        
        elif opcao == 9 :
            F = multiplicarLista(L)
            verFracao(F)
        
        elif opcao == 0 :
            print ("Até à próxima!")
        
        else :
            print ("Opção inválida!")
            
operacoesFracoes()
```
