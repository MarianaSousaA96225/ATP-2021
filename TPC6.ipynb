### Resumo:

Neste ficheiro encontra-se o código fonte, elaborado em linguagem _Python_ , que permite ao utilizador efetuar uma variedade de __operações sobre polinómios__.

Inicialmente, será exibido o menu de opções ao utilizador, com recurso à _função menu()_ , a partir do qual poderá escolher a operação desejada, recursivamente, até escolher a __opção 0__ para __sair__ do programa.

* Ao eligir a __opção 1__, o utilizador poderá __criar__ um polinómio, através da _função criarPolinomio()_ , e de seguida com a __opção 2__ pode __listar__ o seu polinómio, ou seja, vê-lo no output do programa. Para este segundo passo utiliza-se as _funções verTermo() e verPolinomio()_ . 

* Com a __opção 3__ é possível determinar o __grau do polinómio__ gerado com recurso à _função grauPolinomio()_ , com a __opção 4__ é permitido ao utilizador calcular o __valor do polinómio__ num determinado ponto à sua escolha, isto é possível pela _função calcularPolinomio()_ e com a __opção 5__ pode ainda determinar a __derivada do polinómio__ com a _função derivada()_ .

* Ainda tem a escolha de __simplificar o polinómio__, usando para tal a _função simplificar()_ que é chamada quando o utilizador escolher a __opção 6__.

* Adicionou-se a este programa a opção de __criar um gráfico do polinómio__, tal opção não estava previamente disponível no programa elaborado para o __TPC5__. Para a realização do gráfico é chamada a _função grafico()_ na __opção 7__, sendo que esta função requer o import do _matplotlib.pyplot_ - uma coleção de funções que permitem ao _matplotlib_ funcionar como o _MATLAB_ . 



```python
def menu():
    print ("""
    (1) Criar polinómio;
    (2) Listar polinómio;
    (3) Econtrar grau do polinómio;
    (4) Calcular polinómio num ponto;
    (5) Calcular derivada;
    (6) Simplificar polinómio;
    (7) Criar gráfico do polinómio;
    (0) Sair
    """)
    
def criarPolinomio():
    polinomio = []
    n = int(input("quantos termos tem o polinómio?: "))
    j = 1
    for i in range (n) :
        coef = int(input("introduza o coeficiente do termo " + str(j) + ": "))
        exp = int(input("introduza o expoente do termo " + str(j) + ": "))
        j = j + 1
        termo = (coef, exp)
        polinomio.append(termo)
        i = i + 1
    return polinomio
    
def verTermo(termo):
    coef, exp = termo
    if exp == 0 :
        return str(coef)
    elif coef == 1 :
        return "x^" + str(exp)
    else :
        return str(coef) + "x^" + str(exp)
    
def verPolinomio(polinomio):
    politexto = ""
    if len(polinomio) != 0 :
        politexto = verTermo(polinomio[0])
        for i in range (1, len(polinomio)) :
            if polinomio[i][0] >= 0 :
                politexto = politexto + " + " + verTermo(polinomio[i])
            else :
                politexto = politexto + " " + verTermo(polinomio[i])
    print (politexto)

def grauPolinomio(polinomio):
    grau = polinomio[0][1]
    for i in range (1, len(polinomio)) :
        if polinomio[i][1] > grau :
            grau = polinomio[i][1]
    print ("o grau do polinómio é: ", grau)

def calcularPolinomio(polinomio, x):
    valor = 0
    for i in polinomio :
        coef, exp = i
        valor = valor + (coef * (x ** exp))
    return valor

def derivada(polinomio) :
    derivada = []
    for i in polinomio :
        coef, exp = i
        if exp != 0 :
            coef = coef * exp
            exp = exp - 1
            derivadai = (coef, exp)
        else :
            derivadai = (0, 0)
        derivada.append(derivadai)
    return derivada

def ordenarExp(polinomio) :
    trocas = False
    for i in range(len(polinomio)) :
        if i < (len(polinomio) - 1) :
            if polinomio[i][1] < polinomio[i+1][1] :
                polinomio[i], polinomio[i+1] = polinomio[i+1], polinomio[i]
                trocas = True
    if trocas == True :
        (ordenarExp(polinomio))
    return polinomio

def simplificar(polinomioOrd) :
    polinomioSimp = []
    while len(polinomioOrd) > 1 :
        simp = polinomioOrd[0]
        del polinomioOrd[0]
        while simp[1] == polinomioOrd[0][1] :
            simp = (simp[0] + polinomioOrd[0][0], simp[1])
            del polinomioOrd[0]
        polinomioSimp.append(simp)
    if polinomioOrd[0][1] != polinomioSimp[-1][1] :
        polinomioSimp.append(polinomioOrd[0])
    return polinomioSimp
```

```python
import matplotlib.pyplot as plt

def grafico(polinomio) :
    x = [n for n in range(-100, 100)]
    y = []
    for n in x :
        y.append(calcularPolinomio(polinomio, n))
    
    plt.plot(x, y)

    eixoXx = [x[0]-1, x[-1]+1]
    eixoXy = [0, 0]
    plt.plot(eixoXx, eixoXy, color = 'red', linewidth = 1)

    eixoYx = [0, 0]
    eixoYy = [calcularPolinomio(polinomio, x[0]), calcularPolinomio(polinomio, x[-1])]
    plt.plot(eixoYx, eixoYy, color = 'red', linewidth = 1)

    plt.title('Gráfico do polinómio:')

    plt.show()
```

```python
def operacoesPolinomios():
    opcao = -1
    while opcao != 0 :
        menu ()
        opcao = int(input("escolha uma opção: "))
        
        if opcao == 1 :
            P = criarPolinomio()
            
        elif opcao == 2 :
            verPolinomio(P)
            
        elif opcao == 3 :
            grauPolinomio(P)
            
        elif opcao == 4 :
            x = int(input("qual o valor de x?: "))
            print ("o cálculo do polinómio no ponto x = " + str(x) + " é: ", calcularPolinomio(P, x))
            
        elif opcao == 5 :
            P = derivada(P)
            verPolinomio(P)
        
        elif opcao == 6 :
            P = simplificar(ordenarExp(P))
            verPolinomio(P)
            
        elif opcao == 7 :
            grafico(P)
        
        elif opcao == 0 :
            print ("até à próxima!")
            
        else :
            print ("Opção inválida!")
```
